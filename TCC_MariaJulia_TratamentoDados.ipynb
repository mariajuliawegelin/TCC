## 1 - Carregar os dados e bibliotecas de código

```python
#Importar bibliotecas de código:

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Bases extraídas do SNIS em 6/7/2022

# Informações sobre acesso a água e esgotamento sanitário:

df = pd.read_csv('https://raw.githubusercontent.com/mariajuliawegelin/TCC/main/Desagregado_Informacoes_Agua_Esgoto.csv', decimal=',', sep=';', encoding= 'latin-1')
df.head(5)
df = df[:-1]

df.columns = df.columns.str.replace(' ', '') 

df.tail(5)
```

```python
#Acrescentar Região
def determina_regiao(Estado):
    if Estado in ("DF" , "MS" , "MT" , "GO"):
        return "CO"
    if Estado in ( "AC", 'AP' , "RO" , "PA",'AM','TO','RR'):
        return "N"
    if Estado in ('AL', "BA" , 'CE', "MA" , "PB",'PE' , "PI" , "RN", "SE"):
        return "NE"
    if Estado in ( "RS" , "PR" , "SC"):
        return "S"
    if Estado in ( "SP" , "ES" , "MG" , "RJ"):
        return "SE"

df["REGIAO"] = df.Estado.map(determina_regiao)
```

```python
# Calcular o % com acesso a AG e ES:

df['PERCENT_AG'] = df['POP_TOTAL_AGUA']/df['POP_TOTAL']
df['PERCENT_ES'] = df['POP_TOTAL_ESGOTO']/df['POP_TOTAL']
```

```python
# Retirar colunas que não serão objeto de análise:

df = df.drop(columns=['CódigodoPrestador','Prestador', 'SigladoPrestador', 'VOL_AGUA_CONSUMIDO', 'CONSUMO-EM_ELETRICA_SIST_AGUA', 'VOLUME_ES_COLETADO', 'VOLUME_ES_TRATADO', 'CONSUMO-EM_ELETRICA_SIST_ES' ])
```

```python
# Calcular o incremento anual de acesso a AG e ES entre 2011 e 2020:

df = df.sort_values(['CódigodoMunicípio', 'AnodeReferência'])

# Calcular o crescimento anual:
df['DIFF_AG_YoY'] = df['PERCENT_AG'].diff()
df['DIFF_ES_YoY'] = df['PERCENT_ES'].diff()

df.head()
```

```python
# Informações sobre investimento em água e esgotamento sanitário:

df_investimentos= pd.read_csv('https://raw.githubusercontent.com/mariajuliawegelin/TCC/main/Desagregado_Informacoes_Financeiras.csv', decimal=',', sep=';', encoding= 'latin-1')
df_investimentos = df_investimentos[:-1]
df_investimentos.columns = df_investimentos.columns.str.replace(' ', '') 

# Retirar colunas que não serão objeto de análise:

df_investimentos = df_investimentos.drop(columns=['CódigodoPrestador','Prestador', 'SigladoPrestador' ])

df_investimentos.head()
```

```python
# Carregar base do IGPM acumulado até 2020 ano a ano:

df_igpm = pd.read_csv('https://raw.githubusercontent.com/mariajuliawegelin/TCC/main/IGPM%20-%20IPEA%20DATA.csv', decimal=',', sep=';', encoding= 'utf-8')
df_igpm
```

```python
#Incluir IGP-M acumulado até 2020 na base df_investimentos:

df_investimentos = df_investimentos.merge(df_igpm, left_on = "AnodeReferência", right_on = "Ano", how = "left")
df_investimentos.head()
```

```python
df_investimentos = df_investimentos.drop(['Ano'], axis=1)
```

```python
# Calcular o investimento total no município:

df_investimentos['INVESTIMENTO_TOTAL'] = df_investimentos['INVESTIMENTOS_TOTAIS_PRESTADOR']+df_investimentos['INVESTIMENTOS_TOTAIS_MUNICIPIOS']+df_investimentos['INVESTIMENTOS_TOTAIS_ESTADO']
df_investimentos['INVESTIMENTO_TOTAL_AG'] = df_investimentos['INVESTIMENTO_AGUA_PRESTADOR']+df_investimentos['INVESTIMENTO_AGUA_MUNICIPIOS']+df_investimentos['INVESTIMENTO_AGUA_ESTADO']
df_investimentos['INVESTIMENTO_TOTAL_ES'] = df_investimentos['INVESTIMENTO_ES_PRESTADOR']+df_investimentos['INVESTIMENTO_ES_MUNICIPIOS']+df_investimentos['INVESTIMENTO_ES_ESTADO']


# Atualizar o valor do investimento total pelo IGP-M até 2020:

df_investimentos['INVESTIMENTO_TOTAL_ATUAL2020'] = df_investimentos['INVESTIMENTO_TOTAL']*df_investimentos['igpm_acumulado']
df_investimentos['INVESTIMENTO_TOTAL_AG_ATUAL2020'] = df_investimentos['INVESTIMENTO_TOTAL_AG']*df_investimentos['igpm_acumulado']
df_investimentos['INVESTIMENTO_TOTAL_ES_ATUAL2020'] = df_investimentos['INVESTIMENTO_TOTAL_ES']*df_investimentos['igpm_acumulado']

```

```python
#Juntar os dois dataframes:

df_2 = df.merge(df_investimentos, on = ["CódigodoMunicípio", "Município", "Estado", "AnodeReferência", "Abrangência", "Tipodeserviço", "Naturezajurídica" ], how = "outer")
df_2.info()
```

```python
# Calcular o investimento per capita atualizado até 2020:

df_2['INVESTIMENTO_TOTAL_ATUAL2020_PCP'] = df_2['INVESTIMENTO_TOTAL_ATUAL2020']/df_2['POP_TOTAL']
df_2['INVESTIMENTO_TOTAL_AG_ATUAL2020_PCP'] = df_2['INVESTIMENTO_TOTAL_AG_ATUAL2020']/df_2['POP_TOTAL']
df_2['INVESTIMENTO_TOTAL_ES_ATUAL2020_PCP'] = df_2['INVESTIMENTO_TOTAL_ES_ATUAL2020']/df_2['POP_TOTAL']
```

```python
# Carregar base contendo o idh dos municípios (só tem 2010, como ainda não houve o censo de 2020, ainda não há informações para comparar):

df_idh = pd.read_csv('https://raw.githubusercontent.com/mariajuliawegelin/TCC/main/municipios_ibge_idhm_2010.csv', decimal=',', sep=';', encoding= 'latin-1')

# Carregar base contendo o idh dos municípios (só tem 2010, como ainda não houve o censo de 2020, ainda não há informações para comparar):

df_idh_2010 = df_idh[['COD_IBGE','LAT', 'LNG', 'ALT','IDHM', 'IDHM_RENDA', 'IDHM_EDUCACAO',	'IDHM_LONGEVIDADE' ]]

df_idh_2010.head(5)
```

```python
# Calcular o incremento em AG e ES em 10 anos:

anos=[2011,2020]
df_2011_2020 = df_2[df_2.AnodeReferência.isin(anos)]
df_2011_2020 = df_2011_2020.sort_values(['CódigodoMunicípio', 'AnodeReferência'])


# Calcular o YoY growth
df_2011_2020['DIF_AG_2011_2020'] = df_2011_2020['PERCENT_AG'].diff()
df_2011_2020['DIF_ES_2011_2020'] = df_2011_2020['PERCENT_ES'].diff()

ano=[2020]

df_10_anos = df_2011_2020[df_2011_2020.AnodeReferência.isin(ano)]

df_10_anos.head()
```

```python
# Calcular o valor total dos investimentos em AG e ES em 10 anos (2011 a 2020):

df_investimentos_acumulado = df_investimentos.drop(['Município', 'Estado', 'AnodeReferência','Abrangência','Tipodeserviço','Naturezajurídica'], axis=1)
df_investimentos_acumulado = df_investimentos_acumulado.groupby(['CódigodoMunicípio']).sum()
df_investimentos_acumulado.info()
```

```python
# Juntar os dois df do acumulado de 10 anos:

df_10_anos = df_10_anos.merge(df_investimentos_acumulado, on = "CódigodoMunicípio", how = "left")
df_10_anos.info()
```

```python
# Selecionar as variáveis para análise do acumulado de 10 anos:

df_10_anos_selec = df_10_anos[['CódigodoMunicípio', 'Município', 'Estado', 'AnodeReferência', 'Tipodeserviço', 'Naturezajurídica', 'POP_TOTAL', 'POP_TOTAL_ESGOTO', 'POP_TOTAL_AGUA', 'REGIAO', 'DIF_AG_2011_2020', 'DIF_ES_2011_2020', 'INVESTIMENTO_TOTAL_ATUAL2020_y', 'INVESTIMENTO_TOTAL_AG_ATUAL2020_y', 'INVESTIMENTO_TOTAL_ES_ATUAL2020_y', 'INVESTIMENTO_TOTAL_ATUAL2020_PCP', 'INVESTIMENTO_TOTAL_AG_ATUAL2020_PCP', 'INVESTIMENTO_TOTAL_ES_ATUAL2020_PCP' ]]

df_10_anos_selec.head()
```

```python
#Renomear variáveis de investimento:

df_10_anos_selec = df_10_anos_selec.rename(columns = {'INVESTIMENTO_TOTAL_ATUAL2020_y':'INV_TOTAL', 'INVESTIMENTO_TOTAL_AG_ATUAL2020_y':'INV_TOTAL_AG', 'INVESTIMENTO_TOTAL_ES_ATUAL2020_y':'INV_TOTAL_ES', 'INVESTIMENTO_TOTAL_ATUAL2020_PCP': 'INV_TOTAL_PCP', 'INVESTIMENTO_TOTAL_AG_ATUAL2020_PCP' : 'INV_TOTAL_AG_PCP', 'INVESTIMENTO_TOTAL_ES_ATUAL2020_PCP': 'INV_TOTAL_ES_PCP' })


df_10_anos_selec.head()
```

```python
# Incluir LAT, LNG, ALT e IDH2010:

df_3 = df_10_anos_selec.merge(df_idh_2010, left_on = "CódigodoMunicípio", right_on = "COD_IBGE", how = "left")
df_3.head()
```
